Замінити пропущені дані у стовпцях середніми значеннями за стовпчиком (метод fillna)

```python
res = requests.get("https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%80%D0%BE%D0%B4%D0%B6%D1%83%D0%B2%D0%B0%D0%BD%D1%96%D1%81%D1%82%D1%8C_%D0%B2_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D1%96")
soup = BeautifulSoup(res.text, 'html.parser') #if you find any problem with "lxml" then try using "html.parser" instead
tables = soup.find_all("table", class_="wikitable")
table = tables[14] #Table #11
df = pd.read_html(str(table))[0]
df.replace('...', np.nan, inplace=True)
#df.drop(['2003','2013'],axis=1,inplace=True)
df.iloc[:,1:df.shape[1]]=df.iloc[:,1:df.shape[1]].astype('float')
columns_with_missing_values = df.columns[df.isna().any()].tolist()
for column in columns_with_missing_values:
    column_mean = df[column].mean(skipna=True).round(1)
    df[column].fillna(column_mean, inplace=True)
```
